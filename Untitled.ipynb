```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
```

```python
df=pd.read_csv('NUVEM_PALAVRAS.csv',sep=';')
```

```python
#eliminar valores ausentes trazer apenas coluna em analise
TEXTO = df.dropna(subset=['TEXTO'],axis=0)['TEXTO']
```

```python
#exemplos leitura linha
display(TEXTO.iloc[90])
display(TEXTO.iloc[100])
display(TEXTO.iloc[200])
display(TEXTO.iloc[321])
display(TEXTO.iloc[500])
```

```python
#concatenar palavras
all_TEXTO = " ".join(s for s in TEXTO)

#stopwords(remover palavra sem muito significado)
stopwords = set(STOPWORDS)
stopwords.update(['da','do','na','DE','ao','os','FALHA'])
#gerar wordcloud

NUVEM = WordCloud(stopwords=stopwords,background_color='black',width=1600,height=800).generate(all_TEXTO)
```

```python
#Criar nuvem Palavras(temos as palavras que mais sairam nas falhas ocorridas,ja indica aten√ß√£o especial para sensores e transportes, poderia ir mais a fundo observando o trecho )
fig, ax = plt.subplots(figsize=(20,10))
ax.imshow(NUVEM, interpolation='bilinear')
ax.set_axis_off()
```
